**Random forest classifier for titanic dataset**

```python
import pandas as pd
titanic=pd.read_csv("/content/titanic (1).csv")
titanic
```

```python
titanic=titanic.dropna(subset='Survived')
titanic
x=titanic[['Pclass','Sex','Age','SibSp','Parch','Fare']]
y=titanic['Survived']
```

```python
print(x)
print(y)
```

```python
titanic.dtypes
```

```python
titanic.isnull().sum()
```

```python
import numpy as np
titanic['Age']=titanic['Age'].replace(np.NaN,titanic['Age'].mean())
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x['Sex']=le.fit_transform(x['Sex'])
```

```python
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_pred,y_test)
```

```python
class_report = metrics.classification_report(y_pred,y_test)
```

```python
print("\nclassification report\n",class_report)
```

**Random forest classifier for Identifying hand written digits recognition**

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from sklearn.datasets import load_digits
digit=load_digits()
digit
```

```python
dir(digit)
```

```python
import matplotlib.pyplot as plt
plt.gray()
for i in range(5):
  plt.matshow(digit.images[i])
```

```python
digit.data
```

```python

df=pd.DataFrame(digit.data)
df
```

```python
df['target']=digit.target
df
```

```python
x=df.drop(['target'],axis=1)
y=df['target']
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=70,random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_predicted=model.predict(x_test)
y_predicted
```

```python
from sklearn import metrics
cm=metrics.confusion_matrix(y_predicted,y_test)
cm
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True)
plt.xlabel("predicted")
plt.ylabel("actual")
```

```python
from sklearn import tree
from IPython.display import Image
from sklearn.tree import export_graphviz
import graphviz
```

```python
for i in range(3):
  tree=model.estimators_[i]
  feature_names = [str(name) for name in x_train.columns]
  dot_data=export_graphviz(tree,feature_names=feature_names,
                                filled=True,
                                max_depth=2)
  graph=graphviz.Source(dot_data)
  display(graph)
```

```python
report=metrics.classification_report(y_predicted,y_test)
print(report)
```
